### Continuing from LDA.ipynb: prepare doc2vec input

doc2vec will be done via [NETL-Automatic-Topic-Labelling](https://github.com/sb1992/NETL-Automatic-Topic-Labelling-)

```python
import gensim
```

```python
lda_model = gensim.models.ldamodel.LdaModel.load('lda_model')
# model output has 20 topics

lda_model.show_topic(0, topn=30)
```

```python
import pandas as pd
```

```python
terms = []

for i in range(0, 20):
    terms.append([term[0] for term in lda_model.show_topic(i, topn=30)])
    
terms[:2]
```

```python
columns = []
for i in range(0, 30):
    columns.append('term{}'.format(i))

df = pd.DataFrame(data=terms, columns=columns)
df['topic_id'] = [i for i in range(0, 20)]
df.set_index('topic_id', inplace=True)
df.head()
```

```python
df.to_csv('lda-gensim-output.csv')
```
